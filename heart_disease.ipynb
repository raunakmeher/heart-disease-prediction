```python
#Random Forest
#1. Data gathering
import numpy as np
import pandas as pd
#The Machine Learning algorithm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
#Test train split
from sklearn.model_selection import train_test_split
train=pd.read_csv(r"heart.csv")
print(train.head())
#2.Data preprocessing
print(train.describe())
print(train.isnull().sum())
train_x=train.drop('target',axis=1)
print(train_x.head())
train_y=train['target']
print(train_y.head())

tr_x,cv_x,tr_y,cv_y=train_test_split(train_x,train_y,test_size=30)
print(tr_x.head())
print(tr_y.head())

#3.Choosing and training a model
rf=RandomForestClassifier()
rf.fit(tr_x,tr_y)

#4.Evaluating the model
Accuracy_RandomForest=rf.score(cv_x,cv_y)
print("Accuracy of random forest={}%".format(Accuracy_RandomForest*100))
lgr=LogisticRegression(max_iter=2000)
lgr.fit(tr_x,tr_y)
Accuracy_LogisticRegression=lgr.score(cv_x,cv_y)
print("Accuracy of logistic regression={}%".format(Accuracy_LogisticRegression*100))

#6.predict the data
test=pd.read_csv(r"heart_test.csv")
test_x=test.drop('target',axis=1)
#0-no disease,1-disease
print(test_x.head())
test_x.head()
prd=rf.predict(test_x)
prd
op=test[['age']].copy()
op['target']=prd
print(op.head())
```
